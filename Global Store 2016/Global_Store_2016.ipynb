```python
#importing the modules to use 
import pandas as pd
import numpy as np 
from datetime import datetime 
```

```python
#importing the data into the variable name store
store = pd.read_csv('global_superstore_2016.csv')
```

```python
#Displaying 5 rows for our dataset
store.tail(5)
```

```python
order_date = store['Order Date'].tolist()
order_date[-10:]
```

```python
order_datetime = pd.to_datetime(order_date)
order_datetime
```

```python
ship_date = store['Ship Date'].tolist()
ship_date[-10:]
```

```python
ship_datetime = pd.to_datetime(ship_date)
ship_datetime
```

```python
order_series = pd.Series(order_datetime, name = 'Order-Date')
order_series 
```

```python
ship_series = pd.Series(ship_datetime, name = 'Ship_Date')
ship_series
```

```python
store.columns
```

```python
store_cleaned = store[['Row ID', 'Order ID', 'Order Date', 'Ship Date', 'Ship Mode',
       'Customer ID', 'Customer Name', 'Segment','Category']]
store_cleaned.head(5)
```

```python
store_data = store_con.set_index('Order-Date')
store_data
```
