```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
#assigning the dataframe to titanic name using the pd.read_csv() function 
titanic = pd.read_csv('titanic.csv')
titanic.head(5) #Presenting the first 5 rows of our dataset
```

**The average age of people in the Titanic Ship**

```python
#Creating a groupby object and assigning it to name 'a'
a = titanic.groupby(by = titanic['sex']) 
a
```

```python
#Pulling out the mean of the ages from the dataset and showing the averages by sex since we have earlier on grouped our data by sex
a['age'].mean()
```

**The number of poeple who survived in the titanic**

```python
#Extract the number of people who survived first and then group those survivors by sex. Mind you since sex is not found in titanic['survived'], you will have to call it from titanic da
titanic['survived'].groupby(by = titanic['sex']).sum()
```

**The number od survivors and non survivors based on gender**

```python
titanic['survived'].groupby(by = titanic['sex'])
```

```python
for sex in titanic['survived'].groupby(by = titanic['sex']): 
    if sex == 'male': 
        survivors = titanic['survived' == 1].value_counts()
        non_survivors = titanic['survived' == 0].value_counts()
        print('Male survivors are:', survivors)
        print('Male non - survivors are:', non_survivors)
    else:
        print('null')
```
